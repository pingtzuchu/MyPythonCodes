```python
import xml.etree.ElementTree as ET
import os
ns={"tei":"http://www.tei-c.org/ns/1.0"}
rootdir=r"/Users/pingtzuhu/Google 雲端硬碟/Cloud Drive/程式資料/Oxygen/data/qsw"
def deleteElement(node, ns):
    for child in node.findall("*", ns):
        if child.tag=="{http://www.tei-c.org/ns/1.0}div":
            deleteElement(child, ns)
        elif child.tag=="{http://www.tei-c.org/ns/1.0}head":
            pass
        else:
            node.remove(child)
    return node
newTEI=ET.parse("/Users/pingtzuhu/Google 雲端硬碟/Cloud Drive/程式資料/Oxygen/data/qsw/001.xml")
newroot=newTEI.getroot()
newbody=newroot.find("tei:text/tei:body", ns)
newbody.clear()
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if file[-3:]=="xml":
            print(file)
            tree= ET.parse(os.path.join(subdir, file))
            root= tree.getroot()
            for div in root.findall("tei:text/tei:body/tei:div", ns):
                div = deleteElement(div,ns)
                newbody.append(div)
newTEI.write("/Users/pingtzuhu/Documents/sqw.xml", encoding="utf8")
print("We're done.")

                        
```
